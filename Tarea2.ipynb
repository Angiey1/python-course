### **Calculo de la medida de tendencia central y dispersión a nivel global para el primer y último año incluido en el set de datos**

Se utilizaron las funciones de agregación mean() y std() para los cálculos.

**Primer año**

Media

```python
media_2005 = data[data['anio']==2005]['escalera_vida'].mean()
media_2005
```

Desviación estándar

```python
std_2005 = data[data['anio']==2005]['escalera_vida'].std()
std_2005
```

**Último año**

Media

```python
media_2018 = data[data['anio']==2018]['escalera_vida'].mean()
media_2018
```

Desviación estándar

```python
std_2018 = data[data['anio']==2018]['escalera_vida'].std()
std_2018
```

No obstante, debemos tomar en cuenta que los datos de felicidad de todos los años de todos los países no están incluidos en el set de datos.

A continuación se puede observar la cantidad de datos de países en cada año.

```python
data.groupby('anio')['pais'].count()
```

Por lo que, para hacer una comparación deberemos elegir dos países que tengan datos en el primer y último año.

Sin embargo, se debería considerar que la media y la desviación estándar calculada en esos primeros años del dataset, no corresponde a la realidad de la mayoría de países.

Cantidad de datos (años) por país ordenado de forma descendente

```python
data.groupby('pais')['anio'].count().sort_values(ascending=False)
```

```python
pd.DataFrame(data.groupby('pais')['anio'].count().sort_values(ascending=False))
```

Se eligió Francia y Reino Unido

```python
data_fr = data[data.pais=='Francia'][['pais','anio','escalera_vida','de_escalera_pais_anio','soporte_social']]
```

```python
data_ru = data[data.pais=='Reino Unido'][['pais','anio','escalera_vida','de_escalera_pais_anio','soporte_social']]
```

### **Acceso a la información de la felicidad del primer y último año para los dos países**

**Comparación Media**

```python
pd.DataFrame({'anio':[2005,2018],'media_global':[media_2005,media_2018]})
```

```python
data_fr[data_fr.anio == 2005]['escalera_vida'].values[0]
```

```python
data_test = pd.DataFrame({'anio':[2005,2018],
                          'media_global':[media_2005,media_2018],
                          'felicidad_fr':[data_fr[data_fr.anio == 2005]['escalera_vida'].values[0],data_fr[data_fr.anio == 2018]['escalera_vida'].values[0]],
                          'felicidad_ru':[data_ru[data_ru.anio == 2005]['escalera_vida'].values[0],data_ru[data_ru.anio == 2018]['escalera_vida'].values[0]]})
```

```python
data_test
```

```python
data_test['comparacion_fr'] = data_test.apply(lambda x: (x['media_global']-x['felicidad_fr']),axis=1)
```

```python
data_test['comparacion_ru'] = data_test.apply(lambda x: (x['media_global']-x['felicidad_ru']),axis=1)
```

```python
data_test
```

**Interpretación:**

* En el año 2005 (primer año), la población de Francia es más feliz que la media global, y más feliz que en Reino Unido.
* Reino Unido es más feliz que la media global, pero menos feliz que Francia.
* En el año 2018 (último año), la población de Francia es más feliz que la media global, pero menos feliz que Reino Unido.
* Reino Unido es más feliz que la media global y más feliz que Francia.
* En el caso de Francia, en el año 2005 era más feliz que en el 2018, mientras que Reino Unido en el 2005 era menos feliz que en el 2018.

**Comparación Desviación estándar**

Uso de la variable "de_escalera_pais_anio", para realizar una comparación con la desviación estándar.

**de_escalera_pais_anio:** Desviación estándar por país y año del puntaje de felicidad o Escalera de Vida

```python
data_std = pd.DataFrame({'anio':[2005,2018],
                          'std_global':[std_2005,std_2018],
                          'std_felicidad_fr':[data_fr[data_fr.anio == 2005]['de_escalera_pais_anio'].values[0],
                                              data_fr[data_fr.anio == 2018]['de_escalera_pais_anio'].values[0]],
                          'std_felicidad_ru':[data_ru[data_ru.anio == 2005]['de_escalera_pais_anio'].values[0],
                                              data_ru[data_ru.anio == 2018]['de_escalera_pais_anio'].values[0]]})
```

```python
data_std
```

```python
data_std['comparacion_fr'] = data_std.apply(lambda x: (x['std_global']-x['std_felicidad_fr']),axis=1)
```

```python
data_std['comparacion_ru'] = data_std.apply(lambda x: (x['std_global']-x['std_felicidad_ru']),axis=1)
```

```python
data_std
```

**Interpretación:**

* En el año 2005 (primer año), la población de Francia es más feliz que el valor promedio, y más feliz que en Reino Unido. El valor de std en este caso es más alto que el RU, por lo que está más lejos de la media.
* En el año 2018 (último año), la población de Francia es más feliz que la media global, pero menos feliz que Reino Unido. El valor de std es más alto en RU porque está mas lejos de la media.
